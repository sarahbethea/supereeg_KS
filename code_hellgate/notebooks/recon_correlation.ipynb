```python
import numpy as np
import matplotlib.pyplot as plt
import supereeg as se
import glob
import os
from scipy import stats
%matplotlib
```

```python
results_path_10 = "/home/josecs/miniconda3/envs/supereeg_env/data/npz_cluster/results_10/raw_recon"

within_files_10 = glob.glob(os.path.join(results_path_10, "*within*"))

all_files_10 = glob.glob(os.path.join(results_path_10, "*"))
across_files_10 = [f for f in all_files_10 if "within" not in os.path.basename(f)]

within_data_10 = np.zeros(len(within_files_10))
for i in range(len(within_files_10)):
    load = np.load(within_files_10[i])
    within_data_10[i] = load["corrs"]

within_10_mean = np.mean(within_data_10)
within_10_std = np.std(within_data_10)
within_10_var = np.var(within_data_10)
'''x_within = np.linspace(min(within_data_10), max(within_data_10), 100)
pdf_within = stats.norm.pdf(x_within,within_10_mean,within_10_std)'''


across_data_10 = np.zeros(len(across_files_10))
for i in range(len(across_files_10)):
    load = np.load(across_files_10[i])
    across_data_10[i] = load["corrs"]

across_10_mean = np.mean(across_data_10)
across_10_std = np.std(across_data_10)
across_10_var = np.var(across_data_10)
'''x_across = np.linspace(min(across_data_10), max(across_data_10), 100)
pdf_across = stats.t.pdf(x_across,*stats.t.fit(across_data_10))'''


fig, ax = plt.subplots(1,figsize=(10,6))

ax.hist(within_data_10,bins=200,density=True,color="grey",alpha=0.7,label=f"Within (μ = {within_10_mean:.3f}, σ = {within_10_std:.3f}, σ² = {within_10_var:.3f})")
ax.axvline(within_10_mean, color="grey", linestyle="--", linewidth=2)
#ax.plot(x_within,pdf_within,color="grey")
ax.hist(across_data_10, density= True,bins=200,color="black",alpha=0.7,label=f"Across (μ = {across_10_mean:.3f}, σ = {across_10_std:.3f}, σ² = {across_10_var:.3f})")
ax.axvline(across_10_mean, color="black", linestyle="--", linewidth=2)
#ax.plot(x_across,pdf_across,color="black")

ax.set_xlabel("Correlation")
ax.set_ylabel("Density")
ax.set_title("Run 1:Lambda = 10")
ax.legend()
ax.grid()
```
