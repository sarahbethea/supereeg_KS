```python
import numpy as np
import matplotlib.pyplot as plt
import supereeg as se
import glob
import os
%matplotlib
```

```python
results_path_10 = "/home/josecs/miniconda3/envs/supereeg_env/data/npz_cluster/results_10/raw_recon"

within_files_10 = glob.glob(os.path.join(results_path_10, "*within*"))

all_files_10 = glob.glob(os.path.join(results_path_10, "*"))
across_files_10 = [f for f in all_files_10 if "within" not in os.path.basename(f)]

within_data_10 = np.zeros(len(within_files_10))
for i in range(len(within_files_10)):
    load = np.load(within_files_10[i])
    within_data_10[i] = load["corrs"]

across_data_10 = np.zeros(len(across_files_10))
for i in range(len(across_files_10)):
    load = np.load(across_files_10[i])
    across_data_10[i] = load["corrs"]


fig, ax = plt.subplots(1,figsize=(10,6))

ax.hist(within_data_10,bins=200,color="grey",alpha=0.7,label="Within")
ax.hist(across_data_10,bins=200,color="black",alpha=0.7,label="Across")

ax.set_xlabel("Correlation")
ax.set_ylabel("Concentration")
ax.set_title("Lambda = 10")
ax.legend()
ax.grid()
```
